```python
markets = k.select_markets_by_ticker(latest_p.symbol)
markets
```

```python
bets = latest_p[["symbol", "date", "age", "myBet"]].merge(
    markets, how="left", left_on="symbol", right_on="ticker_name",
)
```

```python
bets["time_til_close"] = bets.close_date - bets.date.dt.tz_convert(
    tz=bets.close_date.dt.tz,
)
```

```python
bets = bets[bets.status == "active"]
```

```python
bets
```

```python
bets[["symbol", "yes_bid", "no_ask", "yes_ask", "last_price", "myBet"]]
```

```python
bets["buyYes"] = bets.myBet >= bets.yes_ask
bets["buyNo"] = bets.myBet <= bets.yes_bid
bets["buyable"] = bets.buyYes | bets.buyNo
```

```python
bets[
    [
        "symbol",
        "yes_bid",
        "no_ask",
        "yes_ask",
        "last_price",
        "myBet",
        "buyYes",
        "buyNo",
        "buyable",
    ]
]
```

```python
bets = bets[bets.buyable]
bets["myp"] = bets.myBet.where(bets.buyYes, 1 - bets.myBet)
bets["mktp"] = bets.yes_ask.where(bets.buyYes, 1 - bets.yes_bid)
bets[["mktq", "myq"]] = 1 - bets[["mktp", "myp"]]
bets["expected_value"] = (bets.myp * bets.mktq) - (bets.myq * bets.mktp)
```

```python
bets = bets[bets.expected_value > 0.04]
bets[
    [
        "symbol",
        "yes_bid",
        "no_ask",
        "yes_ask",
        "last_price",
        "myBet",
        "buyYes",
        "buyNo",
        "buyable",
        "expected_value",
        "age",
        "time_til_close",
    ]
]
```
